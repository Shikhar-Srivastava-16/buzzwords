```python
import json
from collections import defaultdict
import string
from datetime import datetime

# Load the JSON file
with open('../data/bsky.json', 'r') as file:
    json_list = json.load(file)

# Convert the JSON objects to a set of tuples (tweet, date)
tweets = [(item['tweet'], item['date']) for item in json_list]

date_word_count = defaultdict(lambda: defaultdict(int))

# Define the date format (since you only have date without time, the format will be "%Y-%m-%d")
date_format = "%Y-%m-%d"  # Adjust this if your date format is different (e.g., "2025-03-15")

# Process each tuple
for tweet, date in tweets:
    # Convert the date string to a datetime object (only the date part)
    date_obj = datetime.strptime(date, date_format).date()  # Only keep the date (year, month, day)

    # Clean the tweet: remove punctuation and split into words
    words = tweet.translate(str.maketrans('', '', string.punctuation)).lower().split()

    # Count the occurrences of each word for the given date
    for word in words:
        date_word_count[date_obj][word] += 1

# Convert defaultdict to a regular dictionary (optional)
date_word_count = {date: dict(word_count) for date, word_count in date_word_count.items()}

# Sort dates in chronological order
sorted_date_word_count = dict(sorted(date_word_count.items()))

# Prepare list of dates and words
list_dates = list(sorted_date_word_count.keys())  # Dates are now datetime.date() objects
list_words = []

for date in sorted_date_word_count:
    for word in sorted_date_word_count[date]:
        list_words += [word]

# You can now use `list_dates` (as date objects) and `list_words`
```
